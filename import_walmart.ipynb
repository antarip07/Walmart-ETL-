```python
import pandas as pd
import mysql.connector

# ✅ Load CSV File
df = pd.read_csv("Walmart.csv")

# ✅ Data Cleaning
df['unit_price'] = df['unit_price'].replace('[\$,]', '', regex=True)  # Remove dollar signs
df['unit_price'] = pd.to_numeric(df['unit_price'], errors='coerce')  # Convert to float, set invalid to NaN
df = df.dropna(subset=['unit_price'])  # Drop rows where unit_price is NaN

df['date'] = pd.to_datetime(df['date'], errors='coerce').dt.date  # Convert date column
df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S', errors='coerce').dt.time  # Convert time column
df = df.dropna()  # Drop remaining NaN rows

# ✅ MySQL Connection
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="12345678",
    database="walmart_sales"
)
cursor = conn.cursor()

# ✅ Create Table (if not exists)
create_table_query = """
CREATE TABLE IF NOT EXISTS sales (
    invoice_id VARCHAR(50) PRIMARY KEY,
    Branch VARCHAR(10),
    City VARCHAR(50),
    category VARCHAR(50),
    unit_price DECIMAL(10,2),
    quantity INT,
    date DATE,
    time TIME,
    payment_method VARCHAR(20),
    rating FLOAT,
    profit_margin FLOAT
);
"""
cursor.execute(create_table_query)

# ✅ Insert Data into MySQL
insert_query = """
INSERT INTO sales (invoice_id, Branch, City, category, unit_price, quantity, date, time, payment_method, rating, profit_margin)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s);
"""

for _, row in df.iterrows():
    try:
        cursor.execute(insert_query, tuple(row))
    except mysql.connector.Error as e:
        print(f"❌ Error inserting row {row['invoice_id']}: {e}")

conn.commit()
cursor.close()
conn.close()

print("✅ Data Imported Successfully!")
```
